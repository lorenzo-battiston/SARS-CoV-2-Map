Per gestire il campo Location, divido il campo in 4 nuovi campi


```python
covid["Continent"] = covid["Location"].str.split("/").str.get(0)
```

```python
covid["Country"] = covid["Location"].str.split("/").str.get(1)
```

```python
covid["State/Region"] = covid["Location"].str.split("/").str.get(2)
```

```python
covid["County/District/City/Province"] = covid["Location"].str.split("/").str.get(3)
```

Salvo i dati rilevanti per le query in un dataframe nuovo

```python
df = covid[['Pango lineage', 'Submission date', 'Continent', 'Country', 'State/Region', 'County/District/City/Province']]
df
```

Riordino il dataframe per submission date, volendo si potrebbero far ripartire il conteggio degli indici

```python
df = df.sort_values('Submission date', ascending = True)
df
```

Primo esempio di simulazione di una query su Pango lineage con l'utilizzo della contains e delle regex per evitare problemi con Uppercase

```python
import re

df.loc[df['Pango lineage'].str.contains('B',na=False, regex = True)]
```

Risulta che la contains chiaramente prende in considerazione anche i pango BA perchè estrae tutte le stringe che contengono B in qualsiasi posizione della stringa. Per fare un altro esempio ricerco su B.1

```python
#Provo a calcolare tutte quelle con Pango lineage = B. Se uso questa forma trovo anche BA che non è della famiglia B

#Altrimenti posso specificare meglio
df.loc[df['Pango lineage'].str.contains('B.1',na=False, regex = True)]
```

Al fine di fare una ricerca puntuale e precisa su B.1 devo usare ^ prima della 'B' e $ dopo. In questo modo usando la sintassi delle regex sceglie tutti i pango che incominciano per B.1 e che finiscono per B.1. Quindi quelli che valgono esattamente B.1

```python
df.loc[df['Pango lineage'].str.contains('^B.1$',na=False, case=False, regex = True)]
```

In questo caso provo a fare una query su Italy. Uso contains per evitare problemi di spazi

```python
df.loc[df['Country'].str.contains('Italy')]
```

Esempio di ricerca per: Pango lineage == B.1 e Country == Italy

```python
df.loc[(df['Pango lineage'].str.contains('^B.1$',na=False, case=False, regex = True)) & (df['Country'].str.contains('Italy'))]
```
